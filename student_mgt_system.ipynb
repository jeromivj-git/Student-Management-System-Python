```python
from openpyxl import Workbook, load_workbook

# -------------------------------
# CONSTANT DATA
# -------------------------------
SUBJECTS = ("Tamil", "English", "Maths", "Science", "Social Science")

students = []

# -------------------------------
# PERFORMANCE FUNCTION
# -------------------------------
def calculate_performance(grade):
    if grade == "A":
        return "Excellent"
    elif grade == "B":
        return "Good"
    elif grade == "C":
        return "Average"
    else:
        return "Poor"

# -------------------------------
# ADD STUDENT
# -------------------------------
def add_student():
    try:
        name = input("Enter name: ").strip()

        if not name.replace(" ", "").isalpha():
            print("❌ Name should contain only alphabets!")
            return

        age = int(input("Enter age: "))

        marks = {}
        total = 0

        # INPUT MARKS
        for subject in SUBJECTS:
            mark = int(input(f"Enter marks for {subject} (0-100): "))

            if mark < 0 or mark > 100:
                print("❌ Marks must be between 0 and 100!")
                return

            marks[subject] = mark
            total += mark

        # AVERAGE
        avg = total / len(SUBJECTS)

        # GRADE
        if avg >= 90:
            grade = "A"
        elif avg >= 75:
            grade = "B"
        elif avg >= 60:
            grade = "C"
        elif avg >= 40:
            grade = "D"
        else:
            grade = "F"

        # PERFORMANCE
        performance = calculate_performance(grade)

        student = {
            "name": name,
            "age": age,
            "marks": marks,
            "average": avg,
            "grade": grade,
            "performance": performance
        }

        students.append(student)

        print("✅ Student added successfully!")

    except ValueError:
        print("❌ Invalid input! Please enter numeric values only.")

# -------------------------------
# VIEW STUDENTS
# -------------------------------
def view_students():
    if not students:
        print("No records found.")
        return

    for i, s in enumerate(students, 1):
        print(f"\nStudent {i}")
        print("Name:", s["name"])
        print("Age:", s["age"])
        print("Marks:", s["marks"])
        print("Average:", round(s["average"], 2))
        print("Grade:", s["grade"])
        print("Performance:", s["performance"])

# -------------------------------
# SEARCH STUDENT
# -------------------------------
def search_student():
    name = input("Enter name to search: ").strip().lower()

    for s in students:
        if s["name"].lower() == name:
            print("Found:")
            print(s)
            return

    print("Student not found.")

# -------------------------------
# UPDATE STUDENT
# -------------------------------
def update_student():
    name = input("Enter name to update: ").strip().lower()

    for s in students:
        if s["name"].lower() == name:

            print("\nWhat do you want to update?")
            print("1. Name")
            print("2. Age")
            print("3. Marks")

            choice = input("Enter choice: ")

            # -----------------------
            # UPDATE NAME
            # -----------------------
            if choice == "1":
                new_name = input("Enter new name: ").strip()
                if not new_name.replace(" ", "").isalpha():
                    print("❌ Invalid name!")
                    return
                s["name"] = new_name

            # -----------------------
            # UPDATE AGE
            # -----------------------
            elif choice == "2":
                try:
                    new_age = int(input("Enter new age: "))
                    s["age"] = new_age
                except ValueError:
                    print("❌ Age must be a number!")
                    return

            # -----------------------
            # UPDATE MARKS
            # -----------------------
            elif choice == "3":
                print("\nSelect subject to update:")
                print("t  -> Tamil")
                print("e  -> English")
                print("m  -> Maths")
                print("s  -> Science")
                print("ss -> Social Science")

                sub_choice = input("Enter subject code: ").strip().lower()

                subject_map = {
                    "t": "Tamil",
                    "e": "English",
                    "m": "Maths",
                    "s": "Science",
                    "ss": "Social Science"
                }

                if sub_choice not in subject_map:
                    print("❌ Invalid subject choice!")
                    return

                subject = subject_map[sub_choice]

                try:
                    new_mark = int(input(f"Enter new marks for {subject}: "))

                    if new_mark < 0 or new_mark > 100:
                        print("❌ Marks must be between 0 and 100!")
                        return

                    s["marks"][subject] = new_mark

                except ValueError:
                    print("❌ Invalid input!")
                    return

                # 🔥 Recalculate after updating marks
                total = sum(s["marks"].values())
                avg = total / len(s["marks"])
                s["average"] = avg

                # Grade logic
                if avg >= 90:
                    grade = "A"
                elif avg >= 75:
                    grade = "B"
                elif avg >= 60:
                    grade = "C"
                elif avg >= 40:
                    grade = "D"
                else:
                    grade = "F"

                s["grade"] = grade
                s["performance"] = calculate_performance(grade)

            else:
                print("❌ Invalid choice!")
                return

            print("✅ Student updated successfully!")
            return

    print("Student not found.")
# -------------------------------
# DELETE STUDENT
# -------------------------------
def delete_student():
    name = input("Enter name to delete: ").strip().lower()

    for s in students:
        if s["name"].lower() == name:
            students.remove(s)
            print("🗑️ Deleted successfully!")
            return

    print("Student not found.")

# -------------------------------
# SORT STUDENTS
# -------------------------------
def sort_students():
    sorted_list = sorted(students, key=lambda x: x["average"], reverse=True)

    print("\n--- Sorted by Average ---")
    for s in sorted_list:
        print(s["name"], "-", round(s["average"], 2))

# -------------------------------
# FILTER STUDENTS
# -------------------------------
def filter_students():
    print("\n--- PERFORMANCE SUMMARY ---")

    # COUNTERS
    excellent_count = 0
    good_count = 0
    average_count = 0
    poor_count = 0

    # COUNT LOOP
    for s in students:
        if s["performance"] == "Excellent":
            excellent_count += 1
        elif s["performance"] == "Good":
            good_count += 1
        elif s["performance"] == "Average":
            average_count += 1
        else:
            poor_count += 1

    # DISPLAY SUMMARY
    print("Excellent:", excellent_count)
    print("Good:", good_count)
    print("Average:", average_count)
    print("Poor:", poor_count)

    # USER INPUT FOR FILTER
    print("\nFilter Options: Excellent / Good / Average / Poor")
    choice = input("Enter performance to filter: ").strip().title()

    print(f"\n--- {choice} Students ---")

    found = False

    for s in students:
        if s["performance"] == choice:
            print("\nName:", s["name"])
            print("Age:", s["age"])
            print("Marks:", s["marks"])
            print("Average:", round(s["average"], 2))
            print("Grade:", s["grade"])
            print("Performance:", s["performance"])
            found = True

    if not found:
        print("❌ No students found in this category.")
# -------------------------------
# SAVE TO EXCEL
# -------------------------------
def save_to_excel():
    wb = Workbook()
    ws = wb.active
    ws.title = "Students"

    ws.append(["Name", "Age", "Tamil", "English", "Maths", "Science", "Social Science", "Average", "Grade", "Performance"])

    for s in students:
        m = s["marks"]

        ws.append([
            s["name"],
            s["age"],
            m["Tamil"],
            m["English"],
            m["Maths"],
            m["Science"],
            m["Social Science"],
            s["average"],
            s["grade"],
            s["performance"]
        ])

    wb.save("students.xlsx")
    print("💾 Data saved to Excel!")

# -------------------------------
# LOAD FROM EXCEL
# -------------------------------
def load_from_excel():
    global students

    try:
        wb = load_workbook("students.xlsx")
        ws = wb.active

        students = []

        for row in ws.iter_rows(min_row=2, values_only=True):
            name, age, t, e, m, s, ss, avg, grade, perf = row

            students.append({
                "name": name,
                "age": age,
                "marks": {
                    "Tamil": t,
                    "English": e,
                    "Maths": m,
                    "Science": s,
                    "Social Science": ss
                },
                "average": avg,
                "grade": grade,
                "performance": perf
            })

        print("📂 Data loaded from Excel!")

    except FileNotFoundError:
        print("No Excel file found. Starting fresh.")

# -------------------------------
# MAIN MENU
# -------------------------------
def main():
    load_from_excel()

    while True:
        print("\n===== STUDENT MANAGEMENT SYSTEM =====")
        print("1. Add Student")
        print("2. View Students")
        print("3. Search Student")
        print("4. Update Student")
        print("5. Delete Student")
        print("6. Sort Students")
        print("7. Filter Students")
        print("8. Save to Excel")
        print("9. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            add_student()
        elif choice == "2":
            view_students()
        elif choice == "3":
            search_student()
        elif choice == "4":
            update_student()
        elif choice == "5":
            delete_student()
        elif choice == "6":
            sort_students()
        elif choice == "7":
            filter_students()
        elif choice == "8":
            save_to_excel()
        elif choice == "9":
            save_to_excel()
            print("Exiting...")
            break
        else:
            print("Invalid choice!")

main()
```
